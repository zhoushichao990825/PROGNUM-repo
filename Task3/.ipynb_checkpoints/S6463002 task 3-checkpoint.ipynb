```python
d=float(input("d"))
D=float(input("D"))
from math import *
def angular_size(d,D):
    Delta= degrees(2*atan(d/(2*D)))
    return Delta

print(f"the angular size is :{angular_size(d,D)},degrees")
```

```python
from math import *
r=float(input("r"))
h=float(input("h"))
def cylinder(r,h):
    a=(2*pi*r**2)+(2*pi*r*h)
    v= h*pi*r**2
    return a,v
print (f"the area and volume are :{cylinder(r,h)}")
cylinder(r,h)
area,volume=cylinder(r,h)
print(area/volume)
```
