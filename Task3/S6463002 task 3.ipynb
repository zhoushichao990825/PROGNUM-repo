```python
import numpy as np
input_array=np.array([2,8,6,2,5,3,6,4,8,9,11,13])
def stdev(input_array):
    n=len(input_array)
    Mean=np.sum(input_array)/n
    Variance=(input_array-Mean)**2
    sigma=np.sqrt(np.sum(Variance)/n)
    return sigma
Sigma=stdev(input_array)
print(f"my value:{Sigma}")
print(f"numpy value:{np.std(input_array)}")      
```

```python
import matplotlib.pyplot as plt
x_value=[]
y_value=[]
from math import *
for x in range(-5,6,):
        y=cosh(x)
        x_value.append(x)
        y_value.append(y)    
plt.plot(x_value, y_value, label="Chain") 
plt.xlabel("x", fontsize=14)
plt.ylabel("y", fontsize=14)
plt.title("A plot", fontsize=25)
plt.xticks(range(-5, 6), fontsize=12)
plt.yticks(range(0, 81, 10), fontsize=12) 
plt.grid(True)
plt.legend(fontsize=12)
plt.show()
```
