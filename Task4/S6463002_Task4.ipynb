```python
import numpy as np
rng = np.random.default_rng()
def game(): # Defining the game function
    print("Welcome to the game of Rock,Paper,scissors!")
    print("Instructions: please input R for rock, P for paper and S for scissor") # Game rule
    player_input=str(input("R/P/S")) 
    RNG=rng.integers(low=1, high=4) #Let the computer input the 1-3 from RNG
    if RNG==1:
            computer_input="R"# Assigning each number generated with an answer
    elif RNG==2:
            computer_input="P"
    else:
            computer_input="S"
    print(f"your choise:{player_input}") # Adding context
    print(f"the computer chose:{computer_input}")
    if player_input==computer_input: # Judging the result
            print("Result: Tie") 
    elif player_input=="R" and computer_input=="S":
            print("Result: You Win")
    elif player_input=="P" and computer_input=="R":
            print("Result: You win")
    elif player_input=="S" and computer_input=="P":
            print("Result: You win")
    else:
            print("Result: You lose")
game()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
data=np.random.normal(185,8,10000)
mu=np.mean(data)
sigma=np.std(data)
h = 185
prob_h = norm.cdf(h, loc=mu, scale=sigma)
lower_bound = mu - sigma
upper_bound = mu + sigma
prob_within_1_sd = norm.cdf(upper_bound, loc=mu, scale=sigma) - norm.cdf(lower_bound, loc=mu, scale=sigma)
print(f"Probability of a value <= {h}: {prob_h:.4f}")
print(f"Probability within 1 Standard Deviation: {prob_within_1_sd:.4f}")
plt.hist(data,bins=40,color='cyan', edgecolor='black', 
         label=f'Mean: {mu:.2f}\nStd Dev: {sigma:.2f}')
plt.axvline(mu, color='red', linestyle='--', linewidth=2, label=f'Mean: {mu:.2f}')
plt.axvspan(mu - sigma, mu + sigma, color='yellow', alpha=0.3, label=f'1 Std Dev: {sigma:.2f}')
plt.title("normal distribution")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.show()

```

```python
import matplotlib.pyplot as plt
import math
import numpy as np
x = np.array([1,3,3,1,1])
y = np.array([1,1,3,3,1])
z = x + 1j * y
angle=np.deg2rad(30)
rotor=math.cos(angle)+1j*math.sin(angle)
result=z*rotor
real=np.real(result)
imaginary=np.imag(result)
plt.plot(x,y,label="Original Cube")
plt.plot(real,imaginary,label="Rotated Cube")
plt.title("Rotor")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.show()
```

```python
import scipy
from astropy.cosmology import Planck18
z=float(input("z value"))
H0 = Planck18.H0.to_value("1/Gyr")
Om0 = Planck18.Om0
Ode0 = Planck18.Ode0
def integrand(z):
    subexpression= H0*(Om0*(1+z)**3+Ode0)**0.5
    return 1/((1+z)*subexpression)
lookback_time=scipy.integrate.quad(integrand,0,z)
print(lookback_time)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

np.random.seed(42)

X = np.random.normal(loc=0, scale=1, size=100000)
Y = np.random.normal(loc=2, scale=1, size=100000)

# PLOT 1: 1D Histogram + PDF

fig1, ax1 = plt.subplots(figsize=(8, 6))

# Plot histograms
ax1.hist(X, bins=30, density=True, alpha=0.5, color='blue', edgecolor='black', label='X')
ax1.hist(Y, bins=30, density=True, alpha=0.5, color='red', edgecolor='black', label='Y')

# Calculate and plot the PDFs
x_axis = np.linspace(-6, 8, 1000)
pdf_x = norm.pdf(x_axis, loc=0, scale=1)
pdf_y = norm.pdf(x_axis, loc=2, scale=1)

ax1.plot(x_axis, pdf_x, color='black', linestyle=':', linewidth=4, label='X(PDF)')
ax1.plot(x_axis, pdf_y, color='black', linestyle='--', linewidth=4, label='Y(PDF)')

# --- NEW EXACT ARROW ANNOTATIONS ---
# Blue arrow 
ax1.annotate(r'$\mathcal{N}(0, 1)$', 
             xy=(-2.0, 0.15),          # Point of the arrow
             xytext=(-4.5, 0.15),      # Center of the text
             color='blue', 
             fontsize=18, 
             va='center', 
             ha='center',
             arrowprops=dict(facecolor='blue', edgecolor='blue', 
                             width=7, headwidth=20, headlength=18, shrink=0.08))

# Red arrow 
ax1.annotate(r'$\mathcal{N}(2, 1)$', 
             xy=(4.0, 0.15),           # Point of the arrow
             xytext=(6.5, 0.15),       # Center of the text
             color='red', 
             fontsize=18, 
             va='center', 
             ha='center',
             arrowprops=dict(facecolor='red', edgecolor='red', 
                             width=7, headwidth=20, headlength=18, shrink=0.08))

# Formatting Plot 1
ax1.set_title('1D Histogram + PDF', fontsize=14)
ax1.set_xlabel('Values', fontsize=12)
ax1.set_ylabel('Probability Density', fontsize=12)
ax1.set_xlim([-6, 8])
ax1.set_xticks(np.arange(-6, 9, 1))
ax1.grid(axis='y', linestyle='--', alpha=0.7)
ax1.legend(loc='upper right')

# ==========================================
# 3. PLOT 2: 2D Histogram + Contour
# ==========================================
fig2, ax2 = plt.subplots(figsize=(7, 6))

# Plot 2D Histogram
counts, xedges, yedges, im = ax2.hist2d(X, Y, bins=30, cmap='Blues')
fig2.colorbar(im, ax=ax2)

# Calculate contour levels based on standard deviations (68%, 95%, 99.7%)
xcenters = (xedges[:-1] + xedges[1:]) / 2
ycenters = (yedges[:-1] + yedges[1:]) / 2
X_grid, Y_grid = np.meshgrid(xcenters, ycenters)

# Sort the 2D bins
sorted_counts = np.sort(counts.flatten())[::-1]
cumulative_counts = np.cumsum(sorted_counts)
total_counts = cumulative_counts[-1]

# Find the threshold values
level_1sigma = sorted_counts[np.searchsorted(cumulative_counts, 0.6827 * total_counts)]
level_2sigma = sorted_counts[np.searchsorted(cumulative_counts, 0.9545 * total_counts)]
level_3sigma = sorted_counts[np.searchsorted(cumulative_counts, 0.9973 * total_counts)]

# Plot Contours
levels = [level_3sigma, level_2sigma, level_1sigma]
CS = ax2.contour(X_grid, Y_grid, counts.T, levels=levels, colors='red', linewidths=1.5)

# Label the contour lines
fmt = {level_1sigma: r'1$\sigma$', level_2sigma: r'2$\sigma$', level_3sigma: r'3$\sigma$'}
ax2.clabel(CS, CS.levels, inline=True, fmt=fmt, fontsize=14, colors='red')

# Formatting Plot 2
ax2.set_title('2D Histogram + Contour', fontsize=14)
ax2.set_xlabel('X', fontsize=12)
ax2.set_ylabel('Y', fontsize=12)
ax2.grid(True, linestyle='--', color='black', alpha=0.8)

# Show both plots
plt.tight_layout()
plt.show()
```
