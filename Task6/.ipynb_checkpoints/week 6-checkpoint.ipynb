```python
from astropy.io import fits
import matplotlib.pyplot as plt
import numpy as np
from scipy import stats
from mpl_toolkits.axes_grid1 import make_axes_locatable 

# Open and print the first 5 headers
hdul = fits.open(r"/Users/users/szhou/Downloads/m101.fits")
hdul.info()
print(hdul[0].header[:5])

# Flatten data
dat = hdul[0].data
flat_dat = np.asarray(dat, dtype=np.float64).ravel()

# Statistics
my_mean = np.mean(flat_dat)
my_median = np.median(flat_dat)
my_std = np.std(flat_dat)
my_max = np.max(flat_dat)
my_min = np.min(flat_dat)
mode_result = stats.mode(flat_dat, keepdims=True)
my_mode = mode_result.mode[0]

print(f"Mean: {my_mean}\nStd Dev: {my_std}\nMedian: {my_median}")
print(f"Max: {my_max}\nMin: {my_min}\nMode: {my_mode}")

# Plot Histogram
fig1, ax1 = plt.subplots()
ax1.hist(flat_dat, bins=100)
ax1.axvline(my_mean, color='r', linestyle='--', label='Mean')
ax1.axvline(my_median, color='b', linestyle=':', label='Median')
ax1.axvline(my_mode, color='c', linestyle='-', label='Mode')
ax1.legend()
ax1.set_title("Data Distribution")
plt.show()

fig2, ax2 = plt.subplots(figsize=(8, 8))
im = ax2.imshow(dat, vmin=my_median, cmap='magma') 

divider = make_axes_locatable(ax2)
cax = divider.append_axes("right", size="5%", pad=0.1)
cbar = fig2.colorbar(im, cax=cax)
cbar.set_label('Brightness (Counts)')


label_x = "CTYPE1"
label_y = "CTYPE2"

ax2.set_xlabel(label_x)
ax2.set_ylabel(label_y)
ax2.set_title("M101 Galaxy Image")

plt.tight_layout() 
plt.show()

hdul[0].header['BUNIT'] = 'Counts'
hdul.writeto('m101_processed.fits', overwrite=True, output_verify='silentfix')
hdul.close()
```

```python
from astropy.io import fits
import matplotlib.pyplot as plt
import numpy as np
from scipy import stats
from mpl_toolkits.axes_grid1 import make_axes_locatable 
hdul = fits.open(r"/Users/users/szhou/Downloads/noisemap0.fits")
hdul2 = fits.open(r"/Users/users/szhou/Downloads/noisemap1.fits")
hdul3 = fits.open(r"/Users/users/szhou/Downloads/noisemap2.fits")
dat = hdul[0].data
flat_dat = dat.flatten()
dat2 = hdul2[0].data
flat_dat2 = dat2.flatten()
dat3 = hdul3[0].data
flat_dat3 = dat3.flatten()
print(f"Standard Deviation for plot 0: {np.std(flat_dat)}")
print(f"Standard Deviation for plot 1: {np.std(flat_dat2)}")
print(f"Standard Deviation for plot 2: {np.std(flat_dat3)}")
stacked_cube = np.stack((dat, dat2, dat3), axis=0)
print(f"\nShape of our new 3D stacked cube: {stacked_cube.shape}")
mean_image = np.mean(stacked_cube, axis=0)
print(f"Shape of the resulting mean image: {mean_image.shape}")
std_mean_image = np.std(mean_image)
print(f"Standard Deviation of the actual MEAN image: {std_mean_image:.4f}")
sigma_0 = np.std(flat_dat)
sigma_1 = np.std(flat_dat2)
sigma_2 = np.std(flat_dat3)
N = 3
# Apply the formula
theoretical_sigma = (1 / N) * np.sqrt(sigma_0**2 + sigma_1**2 + sigma_2**2)
print(f"Theoretical Standard Deviation :    {theoretical_sigma:.4f}")
```
