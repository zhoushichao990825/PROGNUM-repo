```python
from astropy.io import fits
import matplotlib.pyplot as plt
import numpy as np
from scipy import stats
from mpl_toolkits.axes_grid1 import make_axes_locatable 

# Open and print the first 5 headers
hdul = fits.open(r'C:\Users\Gebruiker\Downloads\m101.fits')
hdul.info()
print(hdul[0].header[:5])

# Flatten data
dat = hdul[0].data
flat_dat = np.asarray(dat, dtype=np.float64).ravel()

# Statistics
my_mean = np.mean(flat_dat)
my_median = np.median(flat_dat)
my_std = np.std(flat_dat)
my_max = np.max(flat_dat)
my_min = np.min(flat_dat)
mode_result = stats.mode(flat_dat, keepdims=True)
my_mode = mode_result.mode[0]

print(f"Mean: {my_mean}\nStd Dev: {my_std}\nMedian: {my_median}")
print(f"Max: {my_max}\nMin: {my_min}\nMode: {my_mode}")

# Plot Histogram
fig1, ax1 = plt.subplots()
ax1.hist(flat_dat, bins=100)
ax1.axvline(my_mean, color='r', linestyle='--', label='Mean')
ax1.axvline(my_median, color='b', linestyle=':', label='Median')
ax1.axvline(my_mode, color='c', linestyle='-', label='Mode')
ax1.legend()
ax1.set_title("Data Distribution")
plt.show()

fig2, ax2 = plt.subplots(figsize=(8, 8))
im = ax2.imshow(dat, vmin=my_median, cmap='magma') 

divider = make_axes_locatable(ax2)
cax = divider.append_axes("right", size="5%", pad=0.1)
cbar = fig2.colorbar(im, cax=cax)
cbar.set_label('Brightness (Counts)')


label_x = "CTYPE1"
label_y = "CTYPE2"

ax2.set_xlabel(label_x)
ax2.set_ylabel(label_y)
ax2.set_title("M101 Galaxy Image")

plt.tight_layout() 
plt.show()

hdul[0].header['BUNIT'] = 'Counts'
hdul.writeto('m101_processed.fits', overwrite=True, output_verify='silentfix')
hdul.close()
```

```python
from astropy.io import fits
import matplotlib.pyplot as plt
import numpy as np
from scipy import stats
from mpl_toolkits.axes_grid1 import make_axes_locatable 
hdul = fits.open(r"C:\Users\Gebruiker\Downloads\noisemap0.fits")
hdul2 = fits.open(r"C:\Users\Gebruiker\Downloads/noisemap1.fits")
hdul3 = fits.open(r"C:\Users\Gebruiker\Downloads/noisemap2.fits")
dat = hdul[0].data
flat_dat = dat.flatten()
dat2 = hdul2[0].data
flat_dat2 = dat2.flatten()
dat3 = hdul3[0].data
flat_dat3 = dat3.flatten()
print(f"Standard Deviation for plot 0: {np.std(flat_dat)}")
print(f"Standard Deviation for plot 1: {np.std(flat_dat2)}")
print(f"Standard Deviation for plot 2: {np.std(flat_dat3)}")
stacked_cube = np.stack((dat, dat2, dat3), axis=0)
print(f"\nShape of our new 3D stacked cube: {stacked_cube.shape}")
mean_image = np.mean(stacked_cube, axis=0)
print(f"Shape of the resulting mean image: {mean_image.shape}")
std_mean_image = np.std(mean_image)
print(f"Standard Deviation of the actual MEAN image: {std_mean_image:.4f}")
sigma_0 = np.std(flat_dat)
sigma_1 = np.std(flat_dat2)
sigma_2 = np.std(flat_dat3)
N = 3
# Apply the formula
theoretical_sigma = (1 / N) * np.sqrt(sigma_0**2 + sigma_1**2 + sigma_2**2)
print(f"Theoretical Standard Deviation :    {theoretical_sigma:.4f}")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from astropy.io import fits
from astropy.table import Table, Column
with fits.open(r'C:\Users\Gebruiker\Downloads\APOGEE_stars .fits') as hdul:
    data = Table(hdul[1].data)
parallax_mas = data['GAIAEDR3_PARALLAX'].copy()
parallax_mas[parallax_mas <= 0] = np.nan 
parallax_arcsec = parallax_mas / 1000.0

abs_mag = data['GAIAEDR3_PHOT_G_MEAN_MAG'] + 5 + 5 * np.log10(parallax_arcsec)

if 'DATAFLAG' not in data.colnames:
    data['DATAFLAG'] = 1 

# Apply the specific -1.0 condition
abs_mag[data['DATAFLAG'] == 0] = -1.0

# Add it as a new column to the table
if 'ABS_MAG' in data.colnames:
    data['ABS_MAG'] = abs_mag
else:
    data.add_column(Column(name='ABS_MAG', data=abs_mag))

data.write('APOGEE_update.fits', format='fits', overwrite=True)
print("Updated file saved as 'APOGEE_update.fits'")

#  Plot a figure of absolute magnitude vs color (BP-RP)
# Filter the data so we don't plot the artificial -1.0 values or stars with invalid parallaxes.
plot_mask = (data['DATAFLAG'] != 0) & (data['GAIAEDR3_PARALLAX'] > 0)
plot_data = data[plot_mask]

# Define the X and Y axes for the plot
color_bp_rp = plot_data['GAIAEDR3_PHOT_BP_MEAN_MAG'] - plot_data['GAIAEDR3_PHOT_RP_MEAN_MAG']
plot_abs_mag = plot_data['ABS_MAG']
plt.figure(figsize=(8, 10))
plt.scatter(color_bp_rp, plot_abs_mag, s=0.5, color='black', alpha=0.5)

# Invert Y-axis so brighter stars are at the top 
plt.gca().invert_yaxis() 
plt.xlabel('Color (BP - RP)')
plt.ylabel('Absolute Magnitude ($M_G$)')
plt.title('Color-Magnitude Diagram (Gaia EDR3)')
plt.grid(True, linestyle='--', alpha=0.6)

plt.show()
```

```python
from astroquery.gaia import Gaia

# 1. Define the ADQL query to fetch the requested columns and apply filters
adql_query = """
SELECT TOP 50 
    source_id, ra, dec, parallax, parallax_error, 
    phot_g_mean_mag, phot_bp_mean_mag, phot_rp_mean_mag
FROM gaiadr3.gaia_source
WHERE parallax / parallax_error > 5
  AND phot_g_mean_mag IS NOT NULL
  AND phot_bp_mean_mag IS NOT NULL
  AND phot_rp_mean_mag IS NOT NULL
ORDER BY random_index
"""

print("Fetching data from Gaia...")
job = Gaia.launch_job(adql_query)
results = job.get_results() # This returns an Astropy Table object

df = results.to_pandas()

print("\nFirst 5 rows of the Pandas DataFrame:")
print(df.head())
results.write("astroquery_gaia.fits", format="fits", overwrite=True)
print("\nData successfully saved to 'astroquery_gaia.fits'")
```

```python
class Fibonacci:
    """Class for calculating Fibonacci sequence"""
    
    def get_nth_term(self, n):
        """
        1. Returns the N-th term Fibonacci number.
        Assuming 0-based indexing: 0th term is 0, 1st is 1, 2nd is 1, 3rd is 2...
        """
        if n < 0:
            return "N must be a non-negative integer."
        if n == 0:
            return 0
            
        a, b = 0, 1
        for _ in range(1, n):
            a, b = b, a + b
        return b

    def get_divisible_less_than_nth(self, n, m):
        """
        2. Returns all Fibonacci numbers less than the N-th term 
        that can be divided by M.
        """
        nth_term_value = self.get_nth_term(n)
        divisible_numbers = []
        
        a, b = 0, 1
        # Generate the sequence up until we hit the N-th term
        while a < nth_term_value:
            # Check if the current number a is evenly divisible by m
            if a % m == 0:
                divisible_numbers.append(a)
            # Move to the next sequence numbers
            a, b = b, a + b
            
        return divisible_numbers

# Test for N=100, M=7 

# Instantiate the object
fib_calculator = Fibonacci()

N = 100
M = 7

# 1. Get the 100th term
nth_term = fib_calculator.get_nth_term(N)
print(f"The {N}-th Fibonacci term is:\n{nth_term}\n")

# 2. Get the divisible numbers
divisible_list = fib_calculator.get_divisible_less_than_nth(N, M)
print(f"Fibonacci numbers less than the {N}-th term that are divisible by {M}:")
print(divisible_list)
```
